```
from style_transfer import StyleTransfer
%matplotlib inline
```

```
ST = StyleTransfer()
ST.load_model()
ST.load_content("humberto.jpeg")
ST.load_style("The_Underwave_off_Kanagawa.jpg")

ST.show_images()
```

```
style_weights = {'conv1_1': 1.,
                 'conv2_1': 0.8,
                 'conv3_1': 0.5,
                 'conv4_1': 0.3,
                 'conv5_1': 0.1}

ST.transfer(style_weights, learning_rate=0.003, style_weight=1e8)
```
